```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# Display the first few rows to verify the data
print("First few rows of the dataset:")
print(Iris.head())

# Display all column names
print("\nColumn names in the dataset:")
print(Iris.columns)

# Count the number of examples for each species
print("\nCount of each species:")
species_counts = Iris['Species'].value_counts()
print(species_counts)

# Display unique species in the dataset
print("\nUnique species in the dataset:")
unique_species = Iris['Species'].unique()
print(unique_species)

# Scatter plot of SepalLengthCm vs. SepalWidthCm using the Pandas .plot method
print("\nScatter plot of SepalLengthCm vs. SepalWidthCm:")
Iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm", figsize=(8, 6), color='blue')
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# We can also use the seaborn library to make a similar plot
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(
    data=Iris,
    x="SepalLengthCm",
    y="SepalWidthCm",
    kind="scatter",
    height=8
)

# Add a title to the figure
plt.suptitle("Seaborn Jointplot of Sepal Length vs Sepal Width", y=1.02)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# Use seaborn's FacetGrid to color the scatterplot by species
g = sns.FacetGrid(Iris, hue="Species", height=8)
g.map_dataframe(sns.scatterplot, x="SepalLengthCm", y="SepalWidthCm")
g.add_legend()

# Add a title to the plot
plt.title("Scatter Plot of Sepal Length vs Sepal Width Colored by Species", fontsize=16, pad=20)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# We can look at an individual feature in Seaborn through a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=Iris, x="Species", y="SepalLengthCm")

# Add a title and labels to the plot
plt.title("Boxplot of Sepal Length by Species", fontsize=16, pad=20)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# Create the plot with both a boxplot and a striplot
plt.figure(figsize=(8, 6))

# Boxplot
ax = sns.boxplot(data=Iris, x="Species", y="SepalLengthCm", width=0.6)

# Striplot (adds individual points with jitter)
ax = sns.stripplot(data=Iris, x="Species", y="SepalLengthCm",
                   jitter=True, color="black", size=5, alpha=0.7)

# Add a title and labels
plt.title("Boxplot with Individual Points (Stripplot) of Sepal Length by Species", fontsize=14, pad=20)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# Create a violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=Iris, x="Species", y="SepalLengthCm", inner="quartile", palette="muted")

# Add a title and labels
plt.title("Violin Plot of Sepal Length by Species", fontsize=16, pad=20)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Sepal Length (cm)", fontsize=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# Create a KDE plot for SepalLengthCm by Species
plt.figure(figsize=(8, 6))

# Plot KDE for each species
sns.kdeplot(data=Iris, x="SepalLengthCm", hue="Species", fill=True, common_norm=False, alpha=0.5, palette="muted")

# Add a title and labels
plt.title("KDE Plot of Sepal Length by Species", fontsize=16, pad=20)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Density", fontsize=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# Create a boxplot for each feature split by species
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for feature in features:
    Iris.boxplot(column=feature, by='Species', figsize=(8, 6))
    plt.title(f"Boxplot of {feature} by Species")
    plt.suptitle('')  # Remove default Pandas subtitle
    plt.xlabel("Species")
    plt.ylabel(feature)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# Andrews Curves plot
plt.figure(figsize=(10, 6))
andrews_curves(Iris, "Species", colormap="viridis")

# Add title and labels
plt.title("Andrews Curves of Iris Dataset", fontsize=16)
plt.xlabel("Fourier Series Coefficients", fontsize=12)
plt.ylabel("Value", fontsize=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

# Load the Iris dataset
Iris = pd.read_csv('Iris.csv')

# Parallel coordinates plot
plt.figure(figsize=(12, 6))
parallel_coordinates(Iris, "Species", colormap="viridis")

# Add title and labels
plt.title("Parallel Coordinates Plot of Iris Dataset", fontsize=16)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Feature Values", fontsize=12)

# Show the plot
plt.show()
```
